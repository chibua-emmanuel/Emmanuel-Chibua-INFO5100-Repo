```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
df = pd.read_csv('WineQT.csv')
```

```python
# Display the first few rows of the DataFrame
print(df.head())
```

```python
df.shape
```

```python
# Preprocessing: Check for missing values
print(df.isnull().sum())
```

```python
# Histogram of fixed acidity
plt.figure(figsize=(10,6))
sns.histplot(df['fixed acidity'], kde=False, bins=30)
plt.title('Histogram of Fixed Acidity')
plt.show()
```

```python
# Plotting graphs

# 1. Histogram for 'fixed acidity'
plt.figure(figsize=(6, 4))
sns.histplot(df['fixed acidity'], kde=True, color='blue')
plt.title('Fixed Acidity Distribution')
plt.show()
```

```python
# 2. Boxplot for 'volatile acidity'
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['volatile acidity'], color='green')
plt.title('Volatile Acidity Distribution')
plt.show()
```

```python
# 3. Scatter plot for 'free sulfur dioxide' vs 'total sulfur dioxide'
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['free sulfur dioxide'], y=df['total sulfur dioxide'], color='red')
plt.title('Free Sulfur Dioxide vs Total Sulfur Dioxide')
plt.show()
```

```python
# 4. Correlation heatmap for all features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
